```python
import random
import time
class Kumanda():

    def __init__(self,tv_durum = "Kapalı",tv_ses = 0,kanal_listesi = ["Trt1"],kanal = "Trt1"):
        print("Televizyon Aktif ")
        self.tv_durum = tv_durum
        self.tv_ses = tv_ses
        self.kanal_listesi = kanal_listesi
        self.kanal = kanal

    def tv_ac(self):
        if (self.tv_durum == "Açık"):
            time.sleep(1)
            print("Televizyon Zaten Açık ")
        else:
            print("Televizyon Açılıyor..")
            time.sleep(1)
            self.tv_durum = "Açık"
            print("Televizyon Açıldı")
            time.sleep(1)

    def tv_kapa(self):
        if (self.tv_durum == "Kapalı"):
            time.sleep(1)
            self.tv_ses = 0
            print("Televizyon Zaten Kapalı ")
        else:
            print("Televizyon Kapanıyor..")
            time.sleep(1)
            self.tv_durum = "Kapalı"
            print("Televizyon Kapandı")
            time.sleep(1)


    def ses_ayarlari(self):

        while True:
            time.sleep(1)
            cevap = input("\nSesi Arttır: '>'\nSesi Azalt: '<'\nÇıkış: 'x'\n\nSeçim Yapınız: ")

            if (cevap == ">"):
                if (self.tv_ses != 100):
                    self.tv_ses += 1
                    time.sleep(0.5)
                    print("Ses: ",self.tv_ses)
                else:
                    print("\nMaksimum Ses Sınırına Ulaşıldı")
                    time.sleep(0.5)

            elif (cevap == "<"):
                if (self.tv_ses != 0):
                    self.tv_ses -= 1
                    time.sleep(0.5)
                    print("Ses: ",self.tv_ses)
                else:
                    print("\nMinimum Ses Sınırına Ulaşıldı")
                    time.sleep(0.5)

            elif (cevap == "x" or cevap == "X"):
                print("\nSes Ayarlarından Çıkış Yapılıyor..\n")
                time.sleep(1)
                break

            else:
                print("\nGeçersiz İfade! ")
                time.sleep(1)
                continue

    def kanal_ekle(self,kanal_ismi):

        print("Kanal Ekleniyor..")
        time.sleep(1)

        self.kanal_listesi.append(kanal_ismi)

        print("Kanal Eklendi ")
        time.sleep(1)


    def rastgele_kanal(self):

        rastgele = random.randint(0,len(self.kanal_listesi)-1)

        self.kanal = self.kanal_listesi[rastgele]
        print(self.kanal," Kanalı Açıldı")



    def __len__(self):
        return len(self.kanal_listesi)

    def __str__(self):
        return "\nTv Durumu: {}\nTv Ses: {}\nKanal Listesi: {}\nŞuanki Kanal: {}\n".format(self.tv_durum,self.tv_ses,self.kanal_listesi,self.kanal)


kumanda = Kumanda()

print("""

Televizyon Uygulaması

1.Tv Aç

2.Tv Kapa

3.Ses Ayarları

4.Kanal Ekle

5.Rastgele Kanal

6.Kanal Sayısı Öğrenme

7.Televizyon Bilgileri

8.Programdan Çıkış

""")

while True:
    time.sleep(0.6)
    sec = input("Yapacağınız İşlemi Seçiniz: ")

    if sec == "1":
        kumanda.tv_ac()

    elif sec == "2":
        kumanda.tv_kapa()

    elif sec == "3":
        if kumanda.tv_durum == "Açık":
            kumanda.ses_ayarlari()
        else:
            time.sleep(1)
            print("Televizyon Kapalıyken Ses Ayarlarını Kullanamazsınız ")


    elif sec == "4":
        if kumanda.tv_durum == "Açık":
            kanal_isimleri = input("Ekleyeceğiniz Kanal İsimlerinin Yanına ',' girip kaydediniz. ")
            kanal_listesi = kanal_isimleri.split(",")
            for eklenen_kanallar in kanal_listesi:
                kumanda.kanal_ekle(eklenen_kanallar)
        else:
            time.sleep(1)
            print("Televizyon Kapalıyken Kanal Ekleyemezsiniz ")




    elif sec == "5":
        if kumanda.tv_durum == "Açık":
            kumanda.rastgele_kanal()
        else:
            print("Televizyon Kapalıyken Kanal Değiştiremezsiniz ")

    elif sec == "6":
        if kumanda.tv_durum == "Açık":
            print("Kanal Sayısı: ",len(kumanda))
        else:
            print("Televizyon Kapalıyken Kanal Sayısı Öğrenemezsiniz ")

    elif sec == "7":
        if kumanda.tv_durum == "Açık":
            print(kumanda)
        else:
            print("Televizyon Kapalıyken Televizyon Bilgilerini Görüntüleyemezsiniz ")

    elif sec == "8":
        quit()
    else:
        print("Geçersiz Tuşlama ")
```
